### show all columns of the data/cleaned/*.csv

```python
import os
import pandas as pd

def print_all_column_names(cleaned_folder="data/cleaned"):
    print(f"ğŸ” Lese Spalten aus CSV-Dateien in '{cleaned_folder}'...\n")

    for filename in os.listdir(cleaned_folder):
        if filename.endswith(".csv"):
            filepath = os.path.join(cleaned_folder, filename)
            try:
                df = pd.read_csv(filepath, sep=";")
                print(f"ğŸ“ Datei: {filename}")
                print(f"ğŸ§¾ Spalten: {df.columns.tolist()}\n")
            except Exception as e:
                print(f"âŒ Fehler beim Lesen von '{filename}': {e}\n")
print_all_column_names()
```
